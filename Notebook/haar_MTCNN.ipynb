```python
import cv2
from mtcnn import MTCNN
import matplotlib.pyplot as plt

# Charger l'image
image_path = "DrdPJ0341gUmaF3njuZ3.png"  #  remplace par ton image
img = cv2.imread(image_path)

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# --- Mthode 1 : Haar Cascade ---
haarcascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
faces_haar = haarcascade.detectMultiScale(img, scaleFactor=1.1, minNeighbors=5)

img_haar = img_rgb.copy()
for (x, y, w, h) in faces_haar:
    cv2.rectangle(img_haar, (x, y), (x + w, y + h), (255, 0, 0), 2)  # bleu

# --- Mthode 2 : MTCNN ---
detector = MTCNN()
faces_mtcnn = detector.detect_faces(img_rgb)

img_mtcnn = img_rgb.copy()
for face in faces_mtcnn:
    x, y, w, h = face['box']
    cv2.rectangle(img_mtcnn, (x, y), (x + w, y + h), (0, 255, 0), 2)  # vert

# --- Affichage comparatif ---
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.imshow(img_haar)
plt.title("Haar Cascade Detection")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(img_mtcnn)
plt.title("MTCNN Detection")
plt.axis("off")

plt.show()
```
